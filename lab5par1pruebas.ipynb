```python
from sklearn.datasets import make_blobs
import numpy as np

# Definimos 10 centros
blob_centers = np.array(
    [[ 0.2,  2.3],
     [-1.5 ,  2.3],
     [-2.8,  1.8],
     [-2.8,  2.8],
     [-2.8,  1.3],
     [ 1.0, -1.0],
     [ 2.5,  0.5],
     [-1.0, -2.0],
     [ 0.5,  1.0],
     [ 3.0,  2.0]])

# Definimos las desviaciones estándar para cada uno de los 10 centros
blob_std = np.array([0.4, 0.3, 0.1, 0.1, 0.1, 0.2, 0.3, 0.4, 0.2, 0.3])

# Generamos los blobs con los 10 centros
X, y = make_blobs(n_samples=2000, centers=blob_centers,
                  cluster_std=blob_std, random_state=7)
```

```python
import matplotlib.pyplot as plt

# Función para graficar los clusters
def plot_clusters(X, y=None):
    plt.scatter(X[:, 0], X[:, 1], c=y, s=1, cmap='tab10')  # Usamos un colormap que maneja 10 colores
    plt.xlabel("$x_1$", fontsize=14)
    plt.ylabel("$x_2$", fontsize=14, rotation=0)

# Creamos la figura
plt.figure(figsize=(8, 6))
plot_clusters(X, y)  # Incluimos 'y' para colorear los diferentes clusters
plt.show()
```

```python
#código de prueba

from sklearn.cluster import KMeans

k = 10
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Genera un dataset aleatorio con n_clusters centroides bien separados.
El número de clusters está entre 1 y 10, permitiendo la visualización de los grupos, de la manera más clara posible.

n_samples: Número de puntos datos
n_clusters: Número de cluster o centroides
random_state: Semilla para garantizar la reproducibilidad
cluster_std: Desviación estándar de los clusters (controla la dispersión de los puntos)

```python
def generate_data_with_centroids(n_samples=500, n_clusters=5, random_state=42, cluster_std=0.60):

    # Aseguramos que el número de clusters esté entre 1 y 10
    n_clusters = max(1, min(n_clusters, 10))

    # Generar los datos con 'make_blobs' para tener control sobre la distancia entre clusters
    X, y_true = make_blobs(n_samples=n_samples, centers=n_clusters, cluster_std=cluster_std, random_state=random_state)

    # Visualizar los puntos generados
    plt.scatter(X[:, 0], X[:, 1], c=y_true, s=50, cmap='viridis')
    plt.title(f'Dataset generado con {n_clusters} centroides bien separados')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

    return X, y_true
```

```python
n_centroides = np.random.randint(1, 11)  # Número de centroides aleatorio entre 1 y 10
X, y_true = generate_data_with_centroids(n_samples=500, n_clusters=n_centroides)

# Paso 2: Aplicar K-Means
kmeans = KMeans(n_clusters=n_centroides, random_state=42)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Paso 3: Visualización de los clusters encontrados por K-Means
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

# Mostrar los centroides encontrados
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x')
plt.title('Clusters y Centroides obtenidos por K-Means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
