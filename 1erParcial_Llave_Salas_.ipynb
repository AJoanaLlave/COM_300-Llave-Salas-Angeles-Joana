```python
import random

# Suponiendo que nuestro dataset tiene las siguientes columnas: category, price, is_accessory, etc.
# Este es solo un ejemplo de cómo sería el dataset original en forma de diccionario
data_dict = {
    'category': ['Shoes', 'Bags', 'T-Shirts', 'Shoes', 'Gloves'],
    'price': [49.99, 29.99, 19.99, 89.99, 15.99],
    'is_accessory': [0, 1, 0, 0, 1]
}

# Convertimos el diccionario en un DataFrame para manipularlo más fácilmente (puedes omitir esta parte si ya tienes tu dataset en un DataFrame)
data = pd.DataFrame(data_dict)

# Definimos una función para duplicar datos, variando ligeramente los valores numéricos
def duplicate_data(data):
    # Creamos una copia del dataset original
    data_copy = data.copy()

    # Variamos los precios en un rango del ±10% (es decir, 0.9x a 1.1x su valor original)
    data_copy['price'] = data_copy['price'].apply(lambda x: x * (1 + random.uniform(-0.1, 0.1)))

    # Ahora concatenamos el dataset original con el duplicado
    duplicated_data = pd.concat([data, data_copy], ignore_index=True)

    return duplicated_data

# Duplicamos los datos aplicando la función
duplicated_data = duplicate_data(data)

# Mostramos las primeras filas del dataset duplicado
print("Primeras filas del dataset duplicado:")
print(duplicated_data.head(10))
```
